```python
from scipy.stats import mode
# 中位数
def mode_function(df):
    counts = mode(df)
    return counts[0][0]
train = pd.read_csv("../data/train_2016_v2.csv/train_2016_v2.csv")
train_df = train.merge(data, how='left', on='parcelid')
tmp = train_df.groupby('regionidcounty')['logerror'].agg([('max_', np.max), ('min_', np.min), \
                                                    ('mean_',np.mean),('median_',np.median),
                                                    ('std_',np.std),('mode_',mode_function), 'count']).reset_index()
tmp
```

```python
tmp = train_df.groupby('regionidcity')['logerror'].agg([('max_', np.max), ('min_', np.min), \
                                                    ('mean_',np.mean),('median_',np.median),
                                                    ('std_',np.std),('mode_',mode_function), 'count']).reset_index()
tmp
```
